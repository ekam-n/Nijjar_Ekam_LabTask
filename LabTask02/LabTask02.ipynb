Lets try cleaning. First, let's try just dropping rows with missing values.

```python
df_option1 = df.dropna(subset=['A4', 'A7', 'Final_Exam']).copy()

print(f"Original size: {len(df)}")
print(f"After dropping missing: {len(df_option1)}")
print(f"Rows lost: {len(df) - len(df_option1)}")
```

Lets also try filling the missing values with the given column's mean.

```python
df_option2 = df.copy()
df_option2['A4'] = df_option2['A4'].fillna(df_option2['A4'].mean())
df_option2['A7'] = df_option2['A7'].fillna(df_option2['A7'].mean())

print(f"Mean A4 used: {df['A4'].mean():.2f}")
print(f"Mean A7 used: {df['A7'].mean():.2f}")
print(f"Size: {len(df_option2)}")   
```

Lets also try median.

```python
df_option3 = df.copy()
df_option3['A4'] = df_option3['A4'].fillna(df_option3['A4'].median())
df_option3['A7'] = df_option3['A7'].fillna(df_option3['A7'].median())

print(f"Median A4 used: {df['A4'].median():.2f}")
print(f"Median A7 used: {df['A7'].median():.2f}")
print(f"Size: {len(df_option3)}")
```

Lets evaluate with linear regression.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

def evaluate_strategy(df_option, name):
    X = df_option[['A4', 'A7']]
    y = df_option['Final_Exam']

    model = LinearRegression()
    model.fit(X, y)
    predictions = model.predict(X)

    r2 = r2_score(y, predictions)
    mae = mean_absolute_error(y, predictions)

    print(f"{name}:")
    print(f"  Rows used: {len(df_option)}")
    print(f"  R²:        {r2:.4f}")
    print(f"  MAE:       {mae:.4f}")
    print()

evaluate_strategy(df_option1, "Option 1 (Drop)")
evaluate_strategy(df_option2, "Option 2 (Mean)")
evaluate_strategy(df_option3, "Option 3 (Median)")
```

I tested three strategies for handling missing values in A4 and A7, then evaluated each by training a linear regression model predicting Final_Exam from A4 and A7.

The two metrics disagree on a winner, which itself is interesting:

- Dropping (Option 1) wins on MAE because no fabricated values are introduced — predictions are based purely on real data, so they tend to be slightly more accurate on average. The lower R² is partly a side effect of evaluating on a smaller, harder dataset (only 70 rows), and partly because the dropped students may not be missing at random — students who skip assignments may be systematically different.

- Median imputation (Option 3) wins on R² by a small margin. It outperforms mean imputation because the dataset still contains outliers (negative scores and impossibly high scores), and the median is more robust to those than the mean. Filling in 16 rows with the median makes the model fit slightly better overall.

Now lets detect outliers

```python
max_valid = {'A1': 100, 'A2': 115, 'A3': 100, 'A4': 105,
             'A5': 100, 'A6': 110, 'A7': 100}

print("Outliers per assignment:")
for col, max_val in max_valid.items():
    too_low  = (df[col] < 0).sum()
    too_high = (df[col] > max_val).sum()
    print(f"  {col}: {too_low} negative, {too_high} above {max_val}")
```

Lets clip scores. -4 becomes 0 and 120 becomes 100, etc.

```python
df_B = df.copy()
df_B['A4'] = df_B['A4'].clip(lower=0, upper=105)
df_B['A7'] = df_B['A7'].clip(lower=0, upper=100)
df_B['A4'] = df_B['A4'].fillna(df_B['A4'].median())
df_B['A7'] = df_B['A7'].fillna(df_B['A7'].median())
```
