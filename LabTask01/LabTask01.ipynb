```python
import csv, urllib.request, io

url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"

with urllib.request.urlopen(url) as f:
    content = f.read().decode('utf-8')
    reader = csv.DictReader(io.StringIO(content))   # ← use content, not f
    data = list(reader)
```

```python
import time

df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")
# get the time for the task using pandas
start = time.perf_counter()
names_pd = df['Name']
pd_time = time.perf_counter() - start

# get the time using pure python
start = time.perf_counter()
names_py = [row['Name'] for row in data]
py_time = time.perf_counter() - start

print(f"Pandas time:      {pd_time*1000:.4f} ms")
print(f"Pure Python time: {py_time*1000:.4f} ms")
print(f"First 3 names (pandas): {names_pd.head(3).tolist()}")
print(f"First 3 names (python): {names_py[:3]}")
```

```python
# pandas
start = time.perf_counter()
result_pd = df[(df['Fare'] > 100) & (df['Pclass'] == 1)]
pd_time = time.perf_counter() - start

# python
start = time.perf_counter()
result_py = [
    row for row in data
    if float(row['Fare']) > 100 and int(row['Pclass']) == 1
]
py_time = time.perf_counter() - start

print(f"Pandas time:      {pd_time*1000:.4f} ms")
print(f"Pure Python time: {py_time*1000:.4f} ms")
print(f"Pandas matches:   {len(result_pd)}")
print(f"Python matches:   {len(result_py)}")
```

```python
# pandas
start = time.perf_counter()
avg_age = df['Age'].mean()                
df['AgeCentered'] = df['Age'] - avg_age     
pd_time = time.perf_counter() - start

# python
start = time.perf_counter()

# compute average, skipping missing values
ages = [float(row['Age']) for row in data if row['Age'] != '']
avg_age_py = sum(ages) / len(ages)

# subtract from each row, leaving missing as None
for row in data:
    if row['Age'] != '':
        row['AgeCentered'] = float(row['Age']) - avg_age_py
    else:
        row['AgeCentered'] = None

py_time = time.perf_counter() - start

print(f"Pandas average age:      {avg_age:.4f}")
print(f"Pure Python average:     {avg_age_py:.4f}")
print(f"Pandas time:      {pd_time*1000:.4f} ms")
print(f"Pure Python time: {py_time*1000:.4f} ms")

print(f"\nFirst 5 centered ages (pandas):")
print(df['AgeCentered'].head())
print(f"\nFirst 5 centered ages (python):")
for row in data[:5]:
    print(row['AgeCentered'])
```
